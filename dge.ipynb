```python
#Changing the names of columns for ease of understanding.
#[format: Muscle_condition_Replicate]
new_colnames={}
for col in data.columns:
    if col.startswith('Mmus'):
        parts=col.split("_")
        muscle=parts[2]
        condition=parts[3]
        rep=parts[4]
        short_name= f"{muscle}_{condition}_{rep}"
        print(short_name)
        new_colnames[col]=short_name
        data=data.rename(columns=new_colnames)
```

```python
#Making separate dictionaries for the 2 muscles.
SLS={
'FLT':['SLS_FLT_Rep1','SLS_FLT_Rep2','SLS_FLT_Rep3'],
'GC':['SLS_GC_Rep1','SLS_GC_Rep2','SLS_GC_Rep3'],
'VIV':['SLS_VIV_Rep1','SLS_VIV_Rep2','SLS_VIV_Rep3']}

EDL={
    'FLT':['EDL_FLT_Rep1','EDL_FLT_Rep2','EDL_FLT_Rep3'],
    'GC':['EDL_GC_Rep1','EDL_GC_Rep2','EDL_GC_Rep3'],
    'VIV':['EDL_VIV_Rep1','EDL_VIV_Rep2','EDL_VIV_Rep3']}
```

```python
# Remove NaNs from the ENSEMBL column 
no_nan = data.dropna(subset=['ENSEMBL'])
```

```python
filtered = no_nan.groupby('ENSEMBL').mean(numeric_only=True).reset_index()
```

```python
#Calculating mean values of gene expression in each condition for each gene.
for key in SLS:
    filtered[f'SLS_{key}_mean']= filtered[SLS[key]].mean(axis=1)
    print(filtered[f'SLS_{key}_mean'])
```

```python
#Creating box plots of Mean values of each condition
melted= filtered[['SLS_FLT_mean', 'SLS_GC_mean', 'SLS_VIV_mean']].melt(var_name='Condition', value_name='Expression')
sns.boxplot(x='Condition', y='Expression', hue='Condition', palette="viridis", data=melted)
plt.title("Gene Expression Distribution by Condition")
plt.show()
```

log2fc =1 means 2 folds change in gene expression
log2fc ≈ 0.8 → ~1.74 folds change
log2fc ≈ 0.75 → ~1.68 folds change.          ( below 0.5 can be considered insignificant)

```python
#Calculating Log2 fold change:
filtered['SLS_log2fc_flt_gc'] = filtered['SLS_FLT_mean'] - filtered['SLS_GC_mean']
filtered['SLS_log2fc_flt_viv'] = filtered['SLS_FLT_mean'] - filtered['SLS_VIV_mean']
filtered['SLS_log2fc_gc_viv'] = filtered['SLS_GC_mean'] - filtered['SLS_VIV_mean']
a=(filtered['SLS_log2fc_flt_gc'].min(), filtered['SLS_log2fc_flt_gc'].max())
b=(filtered['SLS_log2fc_flt_viv'].min(), filtered['SLS_log2fc_flt_viv'].max())
c=(filtered['SLS_log2fc_gc_viv'].min(), filtered['SLS_log2fc_gc_viv'].max())
print(f'FLT vs GC- min log2fc value: {a[0]}, max log2fc value: {a[1]}' )
print(f'FLT vs VIV- min log2fc value: {b[0]}, max log2fc value: {b[1]}' )
print(f'GC vs VIV- min log2fc value: {b[0]}, max log2fc value: {b[1]}' )
```

```python
#Creating boxplots of log 2 fold change value of compared conditions
plt.figure(figsize=(5,4))
sns.boxplot(data=filtered[['SLS_log2fc_flt_gc', 'SLS_log2fc_flt_viv', 'SLS_log2fc_gc_viv']])
plt.title("Boxplot of Log2 Fold Changes")
plt.xlabel("Comparision")
plt.xticks(ticks=[0, 1, 2], labels=["FLT vs GC", "FLT vs VIV", "GC vs VIV"])
plt.ylabel("Log2 fold change")
plt.show()
```

```python
#Performing Anova and t-tests
from scipy import stats

FLT = filtered[SLS['FLT']].values
GC = filtered[SLS['GC']].values
VIV = filtered[SLS['VIV']].values

f_statistics = []
anova_pvals = []

t_pval_flt_gc = []
t_pval_flt_viv = []
t_pval_gc_viv = []

def anova_and_ttests(i):
    flt = FLT[i]
    gc = GC[i]
    viv = VIV[i]

    # ANOVA
    f_stat, p_value = stats.f_oneway(flt, gc, viv)
    
    # t-tests
    p_flt_gc = stats.ttest_ind(flt, gc, equal_var=False)
    p_flt_viv = stats.ttest_ind(flt, viv, equal_var=False)
    p_gc_viv = stats.ttest_ind(gc, viv, equal_var=False)

    return f_stat, p_value, p_flt_gc, p_flt_viv, p_gc_viv

for i in range(len(filtered)):
    f, p, t1, t2, t3 = anova_and_ttests(i)
    f_statistics.append(f)
    anova_pvals.append(p)
    t_pval_flt_gc.append(t1.pvalue)
    t_pval_flt_viv.append(t2.pvalue)
    t_pval_gc_viv.append(t3.pvalue)
```

```python
#Classifying genes as rejected or accepted based on ANOVA
p_val= np.array(anova_pvals)
significance_level= 0.05
filtered['anova_null_hypothesis']= "Accepted"
filtered.loc[p_val < significance_level, 'anova_null_hypothesis'] = 'Rejected'
print(filtered['ENSEMBL'].groupby(filtered['anova_null_hypothesis']).count())
```

If p-value<0.05 then null hypothesis is rejected, if more than 0.05 then accepted. Rejeted are significant
