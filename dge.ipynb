WE WILL USE ENSEMBL ID FOR DIFFERENTIATION OF THE GENES AS IT IS UNIQUE 

```python
#Changing the names of columns for ease of understanding.
#[format: Muscle_condition_Replicate]
new_colnames={}
for col in data.columns:
    if col.startswith('Mmus'):
        parts=col.split("_")
        muscle=parts[2]
        condition=parts[3]
        rep=parts[4]
        short_name= f"{muscle}_{condition}_{rep}"
        print(short_name)
        new_colnames[col]=short_name
        data=data.rename(columns=new_colnames)
```

```python
#Making separate dictionaries for the 2 muscles.
SLS={
'FLT':['SLS_FLT_Rep1','SLS_FLT_Rep2','SLS_FLT_Rep3'],
'GC':['SLS_GC_Rep1','SLS_GC_Rep2','SLS_GC_Rep3'],
'VIV':['SLS_VIV_Rep1','SLS_VIV_Rep2','SLS_VIV_Rep3']}

EDL={
    'FLT':['EDL_FLT_Rep1','EDL_FLT_Rep2','EDL_FLT_Rep3'],
    'GC':['EDL_GC_Rep1','EDL_GC_Rep2','EDL_GC_Rep3'],
    'VIV':['EDL_VIV_Rep1','EDL_VIV_Rep2','EDL_VIV_Rep3']}
```

```python
# Remove NaNs from the ENSEMBL column 
no_nan = data.dropna(subset=['ENSEMBL'])
```

```python
filtered = no_nan.groupby('ENSEMBL').mean(numeric_only=True).reset_index()
```

```python
#Calculating mean values of gene expression in each condition for each gene.
for key in SLS:
    filtered[f'SLS_{key}_mean']= filtered[SLS[key]].mean(axis=1)
    print(filtered[f'SLS_{key}_mean'])
```

```python
#Creating box plots of Mean values of each condition
melted= filtered[['SLS_FLT_mean', 'SLS_GC_mean', 'SLS_VIV_mean']].melt(var_name='Condition', value_name='Expression')
sns.boxplot(x='Condition', y='Expression', hue='Condition', palette="viridis", data=melted)
plt.title("Gene Expression Distribution by Condition")
plt.show()
```

log2fc =1 means 2 folds change in gene expression
log2fc ≈ 0.8 → ~1.74 folds change
log2fc ≈ 0.75 → ~1.68 folds change.          ( below 0.5 can be considered insignificant)

```python
#Calculating Log2 fold change:
filtered['SLS_log2fc_flt_gc'] = filtered['SLS_FLT_mean'] - filtered['SLS_GC_mean']
filtered['SLS_log2fc_flt_viv'] = filtered['SLS_FLT_mean'] - filtered['SLS_VIV_mean']
filtered['SLS_log2fc_gc_viv'] = filtered['SLS_GC_mean'] - filtered['SLS_VIV_mean']
a=(filtered['SLS_log2fc_flt_gc'].min(), filtered['SLS_log2fc_flt_gc'].max())
b=(filtered['SLS_log2fc_flt_viv'].min(), filtered['SLS_log2fc_flt_viv'].max())
c=(filtered['SLS_log2fc_gc_viv'].min(), filtered['SLS_log2fc_gc_viv'].max())
print(f'FLT vs GC- min log2fc value: {a[0]}, max log2fc value: {a[1]}' )
print(f'FLT vs VIV- min log2fc value: {b[0]}, max log2fc value: {b[1]}' )
print(f'GC vs VIV- min log2fc value: {b[0]}, max log2fc value: {b[1]}' )
```

```python
#Creating boxplots of log 2 fold change value of compared conditions
plt.figure(figsize=(5,4))
sns.boxplot(data=filtered[['SLS_log2fc_flt_gc', 'SLS_log2fc_flt_viv', 'SLS_log2fc_gc_viv']])
plt.title("Boxplot of Log2 Fold Changes")
plt.xlabel("Comparision")
plt.xticks(ticks=[0, 1, 2], labels=["FLT vs GC", "FLT vs VIV", "GC vs VIV"])
plt.ylabel("Log2 fold change")
plt.show()
```

```python
#Performing Anova and t-tests
from scipy import stats

FLT = filtered[SLS['FLT']].values
GC = filtered[SLS['GC']].values
VIV = filtered[SLS['VIV']].values

f_statistics = []
anova_pvals = []

t_pval_flt_gc = []
t_pval_flt_viv = []
t_pval_gc_viv = []

def anova_and_ttests(i):
    flt = FLT[i]
    gc = GC[i]
    viv = VIV[i]

    # ANOVA
    f_stat, p_value = stats.f_oneway(flt, gc, viv)
    
    # t-tests
    p_flt_gc = stats.ttest_ind(flt, gc, equal_var=False)
    p_flt_viv = stats.ttest_ind(flt, viv, equal_var=False)
    p_gc_viv = stats.ttest_ind(gc, viv, equal_var=False)

    return f_stat, p_value, p_flt_gc, p_flt_viv, p_gc_viv

for i in range(len(filtered)):
    f, p, t1, t2, t3 = anova_and_ttests(i)
    f_statistics.append(f)
    anova_pvals.append(p)
    t_pval_flt_gc.append(t1.pvalue)
    t_pval_flt_viv.append(t2.pvalue)
    t_pval_gc_viv.append(t3.pvalue)
```

```python
#Classifying genes as rejected or accepted based on ANOVA
p_val= np.array(anova_pvals)
significance_level= 0.05
filtered['anova_null_hypothesis']= "Accepted"
filtered.loc[p_val < significance_level, 'anova_null_hypothesis'] = 'Rejected'
print(filtered['ENSEMBL'].groupby(filtered['anova_null_hypothesis']).count())
```

If p-value<0.05 then null hypothesis is rejected, if more than 0.05 then accepted. Rejeted are significant

```python
#plotting p-values obtained from ANOVA
plt.figure(figsize=(6,4))
sns.histplot( y= p_val, hue= filtered['anova_null_hypothesis'], palette='colorblind')
plt.axhline( y=0.05, color='red', linestyle= "--", label= 'Significance Level= 0.05')
plt.ylabel(" p-values")
plt.title("P-Value Distribution of Genes with Respect to Anova Null Hypothesis")
plt.figtext(0.5, -0.05, "All genes below significance level are candidates for differential expression across the three conditions.", ha="center", fontsize=10)
plt.legend()
plt.show()
```

```python
significance_level = 0.05

# Assign t-test p-values as new columns
filtered['pval_flt_gc'] = t_pval_flt_gc
filtered['pval_flt_viv'] = t_pval_flt_viv
filtered['pval_gc_viv'] = t_pval_gc_viv

# Add Null Hypothesis Results
filtered['Null_hypothesis_flt_gc'] = "Accepted"
filtered.loc[filtered['pval_flt_gc'] < significance_level, 'Null_hypothesis_flt_gc'] = 'Rejected'
print(filtered['ENSEMBL'].groupby(filtered['Null_hypothesis_flt_gc']).count())

filtered['Null_hypothesis_flt_viv'] = "Accepted"
filtered.loc[filtered['pval_flt_viv'] < significance_level, 'Null_hypothesis_flt_viv'] = 'Rejected'
print(filtered['ENSEMBL'].groupby(filtered['Null_hypothesis_flt_viv']).count())
```

```python
#Plotting p-values obtained from t-test for the 3 conditions
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
sns.histplot( y= filtered['pval_flt_gc'], hue= filtered['Null_hypothesis_flt_gc'], palette='colorblind')
plt.ylabel(" p-values")
plt.title("P-Value Distribution FLT vs GC ")
plt.subplot(1,2,2)
sns.histplot( y= filtered['pval_flt_viv'], hue= filtered['Null_hypothesis_flt_viv'], palette='colorblind')
plt.ylabel(" p-values")
plt.title("P-Value Distribution FLT vs VIV ")
plt.figtext(0.5, -0.05, "All the rejected genes are significant", ha="center", fontsize=10)
plt.show()
```

```python
#-log10(p-value) calculation for volcano plot
filtered['nlpv_flt_gc']= -np.log10(t_pval_flt_gc)
filtered['nlpv_flt_viv']= -np.log10(t_pval_flt_viv)
filtered['nlpv_gc_viv']= -np.log10(t_pval_gc_viv)
```

```python
#Creating Volcano plots to visualize gene expression changes across conditions.
plt.figure(figsize=(18,5))
plt.subplot(1,3,1)
sns.scatterplot(data= filtered, x= filtered['SLS_log2fc_flt_gc'], y= filtered['nlpv_flt_gc'])
plt.axhline(y= -np.log10(0.05), color= "green", linestyle= "--", label= "Significance level= 1.3")
plt.axvline(x= 1, color= "orange", linestyle="--", label= "up-regulated genes")
plt.axvline(x= -1, color= "red", linestyle="--", label= "down-regulated genes")
plt.ylabel("-log10(p_value)")
plt.xlabel("log2 fold change FLT V/S GC")
plt.title("FLT V/S GC")
plt.legend(loc=0, fontsize=8)
plt.subplot(1,3,2)
sns.scatterplot(data= filtered, x= filtered['SLS_log2fc_flt_viv'] , y= filtered['nlpv_flt_viv'])
plt.axhline(y= -np.log10(0.05), color= "green", linestyle= "--", label= "Significance level= 1.3")
plt.axvline(x= 1, color= "orange", linestyle="--", label= "up-regulated genes")
plt.axvline(x= -1, color= "red", linestyle="--", label= "down-regulated genes")
plt.ylabel("-log10(p_value)")
plt.xlabel("log2 fold change FLT V/S VIV")
plt.title("FLT V/S VIV")
plt.legend(loc=0, fontsize=8)
plt.subplot(1,3,3)
sns.scatterplot(data= filtered, x= filtered['SLS_log2fc_gc_viv'] , y= filtered['nlpv_gc_viv'])
plt.axhline(y= -np.log10(0.05), color= "green", linestyle= "--", label= "Significance level= 1.3")
plt.axvline(x= 1, color= "orange", linestyle="--", label= "up-regulated genes")
plt.axvline(x= -1, color= "red", linestyle="--", label= "down-regulated genes")
plt.ylabel("-log10(p_value)")
plt.xlabel("log2 fold change GC V/S VIV")
plt.title("GC V/S VIV")
plt.legend(loc=0, fontsize=8)
plt.suptitle("Volcano Plots of Differential Gene Expression in SLS muscle", fontsize=16, fontweight='bold')
plt.figtext(0.5, -0.05, "All the genes above significance level and have log2 fold change value near +-1 are significant genes and have shown significant changes over different conditions ", ha="center", fontsize=10)
plt.show()
```

p-value < 0.05 → i.e., -log₁₀(p-value) > 1.3

Performing ANOVA and t-test

```python
#Performing Anova and t-tests
from scipy import stats

EDL_FLT= edl_filtered[EDL['FLT']].values
EDL_GC= edl_filtered[EDL['GC']].values
EDL_VIV= edl_filtered[EDL['VIV']].values
EDL_f_statistics=[]
EDL_anova_pvals=[]


EDL_t_pval_flt_gc = []
EDL_t_pval_flt_viv = []
EDL_t_pval_gc_viv = []

def anova_and_ttest(i):
    E_flt= EDL_FLT[i]
    E_gc= EDL_GC[i]
    E_viv= EDL_VIV[i]

    # ANOVA
    edl_f_stat, edl_p_value = stats.f_oneway(E_flt, E_gc, E_viv)
    
    # t-tests
    edl_p_flt_gc = stats.ttest_ind(E_flt, E_gc, equal_var=False)
    edl_p_flt_viv = stats.ttest_ind(E_flt, E_viv, equal_var=False)
    edl_p_gc_viv = stats.ttest_ind(E_gc, E_viv, equal_var=False)

    return edl_f_stat, edl_p_value, edl_p_flt_gc, edl_p_flt_viv, edl_p_gc_viv

for i in range(len(edl_filtered)):
    F, P, T1, T2, T3 = anova_and_ttest(i)
    EDL_f_statistics.append(F)
    EDL_anova_pvals.append(P)
    EDL_t_pval_flt_gc.append(T1.pvalue)
    EDL_t_pval_flt_viv.append(T2.pvalue)
    EDL_t_pval_gc_viv.append(T3.pvalue)
```

```python
#Classifying genes as rejected or accepted based on ANOVA
EDL_p_val= np.array(EDL_anova_pvals)
significance_level= 0.05
edl_filtered['EDL_anova_null_hypothesis']= "Accepted"
edl_filtered.loc[EDL_p_val < significance_level, 'EDL_anova_null_hypothesis'] = 'Rejected'
print(edl_filtered['ENSEMBL'].groupby(edl_filtered['EDL_anova_null_hypothesis']).count())
```

```python
#plotting p-values obtained from ANOVA
plt.figure(figsize=(6,4))
sns.histplot( y= EDL_p_val, hue= edl_filtered['EDL_anova_null_hypothesis'], palette='colorblind')
plt.ylabel(" p-values")
plt.title("P-Value Distribution of Genes with Respect to Anova Null Hypothesis")
plt.show()
```

```python
significance_level = 0.05

# Assign t-test p-values as new columns
edl_filtered['pvalue_flt_gc'] = EDL_t_pval_flt_gc
edl_filtered['pvalue_flt_viv'] = EDL_t_pval_flt_viv
edl_filtered['pvalue_gc_viv'] = EDL_t_pval_gc_viv

# Add Null Hypothesis Results
edl_filtered['nullhypothesis_flt_gc'] = "Accepted"
edl_filtered.loc[edl_filtered['pvalue_flt_gc'] < significance_level, 'nullhypothesis_flt_gc'] = 'Rejected'
print(edl_filtered['ENSEMBL'].groupby(edl_filtered['nullhypothesis_flt_gc']).count())

edl_filtered['nullhypothesis_flt_viv'] = "Accepted"
edl_filtered.loc[edl_filtered['pvalue_flt_viv'] < significance_level, 'nullhypothesis_flt_viv'] = 'Rejected'
print(edl_filtered['ENSEMBL'].groupby(edl_filtered['nullhypothesis_flt_viv']).count())

edl_filtered['nullhypothesis_gc_viv'] = "Accepted"
edl_filtered.loc[edl_filtered['pvalue_gc_viv'] < significance_level, 'nullhypothesis_gc_viv'] = 'Rejected'
print(edl_filtered['ENSEMBL'].groupby(edl_filtered['nullhypothesis_gc_viv']).count())
```

```python
#Plotting p-values obtained from t-test for the 3 conditions
plt.figure(figsize=(15,4))
plt.subplot(1,3,1)
sns.histplot( y= edl_filtered['pvalue_flt_gc'], hue= edl_filtered['nullhypothesis_flt_gc'], palette='colorblind')
plt.ylabel(" p-values")
plt.title("P-Value Distribution FLT vs GC ")
plt.subplot(1,3,2)
sns.histplot( y= edl_filtered['pvalue_flt_viv'], hue= edl_filtered['nullhypothesis_flt_viv'], palette='colorblind')
plt.ylabel(" p-values")
plt.title("P-Value Distribution FLT vs VIV ")
plt.subplot(1,3,3)
sns.histplot( y= edl_filtered['pvalue_gc_viv'], hue= edl_filtered['nullhypothesis_gc_viv'], palette='colorblind')
plt.ylabel(" p-values")
plt.title("P-Value Distribution GC vs VIV ")
plt.figtext(0.5, -0.05, "All the rejected genes are significant", ha="center", fontsize=10)
plt.show()
```

```python
#-log10(p-value) calculation for volcano plot
edl_filtered['neg_lpv_flt_gc']= -np.log10(EDL_t_pval_flt_gc)
edl_filtered['neg_lpv_flt_viv']= -np.log10(EDL_t_pval_flt_viv)
edl_filtered['neg_lpv_gc_viv']= -np.log10(EDL_t_pval_gc_viv)
```

```python
#Creating Volcano plots to visualize gene expression changes across conditions.
plt.figure(figsize=(18,5))
plt.subplot(1,3,1)
sns.scatterplot(data=edl_filtered, x= edl_filtered['edl_log2fc_FLT_GC'] , y= edl_filtered['neg_lpv_flt_gc'])
plt.axhline(y= -np.log10(0.05), color= "green", linestyle= "--", label= "Significance level= 0.05")
plt.ylabel("-log10(p_value)")
plt.xlabel("log2 fold change FLT V/S GC")
plt.axvline(x= 0.75, color= "orange", linestyle="--", label= "up-regulated genes")
plt.axvline(x= -0.75, color= "red", linestyle="--", label= "down-regulated genes")
plt.title("FLT V/S GC")
plt.legend(loc=0, fontsize=8)

plt.subplot(1,3,2)
sns.scatterplot(data= edl_filtered, x= edl_filtered['edl_log2fc_FLT_VIV'] , y= edl_filtered['neg_lpv_flt_viv'])
plt.axhline(y= -np.log10(0.05), color= "green", linestyle= "--", label= "Significance level= 0.05")
plt.ylabel("-log10(p_value)")
plt.xlabel("log2 fold change FLT V/S VIV")
plt.axvline(x= 0.75, color= "orange", linestyle="--", label= "up-regulated genes")
plt.axvline(x= -0.75, color= "red", linestyle="--", label= "down-regulated genes")
plt.title("FLT V/S VIV")
plt.legend(bbox_to_anchor=(0.6,0.9), fontsize=8)

plt.subplot(1,3,3)
sns.scatterplot(data= edl_filtered, x= edl_filtered['edl_log2fc_GC_VIV'] , y= edl_filtered['neg_lpv_gc_viv'])
plt.axhline(y= -np.log10(0.05), color= "green", linestyle= "--", label= "Significance level= 0.05")
plt.ylabel("-log10(p_value)")
plt.xlabel("log2 fold change GC V/S VIV")
plt.axvline(x=0.75, color= "orange", linestyle="--", label= "up-regulated genes")
plt.axvline(x= -0.75, color= "red", linestyle="--", label= "down-regulated genes")
plt.title("GC V/S VIV")
plt.suptitle("Volcano Plots of Differential Gene Expression in EDL muscle", fontsize=16, fontweight='bold')
plt.figtext(0.5, -0.05, "All the genes above significance level and have log2 fold change value near +-1 are significant genes and have shown significant changes over different conditions ", ha="center", fontsize=10)
plt.legend(loc=0, fontsize=8)
plt.show()
```
