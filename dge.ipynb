WE WILL USE ENSEMBL ID FOR DIFFERENTIATION OF THE GENES AS IT IS UNIQUE 

```python
#Changing the names of columns for ease of understanding.
#[format: Muscle_condition_Replicate]
new_colnames={}
for col in data.columns:
    if col.startswith('Mmus'):
        parts=col.split("_")
        muscle=parts[2]
        condition=parts[3]
        rep=parts[4]
        short_name= f"{muscle}_{condition}_{rep}"
        print(short_name)
        new_colnames[col]=short_name
        data=data.rename(columns=new_colnames)
```

```python
#Making separate dictionaries for the 2 muscles.
SLS={
'FLT':['SLS_FLT_Rep1','SLS_FLT_Rep2','SLS_FLT_Rep3'],
'GC':['SLS_GC_Rep1','SLS_GC_Rep2','SLS_GC_Rep3'],
'VIV':['SLS_VIV_Rep1','SLS_VIV_Rep2','SLS_VIV_Rep3']}

EDL={
    'FLT':['EDL_FLT_Rep1','EDL_FLT_Rep2','EDL_FLT_Rep3'],
    'GC':['EDL_GC_Rep1','EDL_GC_Rep2','EDL_GC_Rep3'],
    'VIV':['EDL_VIV_Rep1','EDL_VIV_Rep2','EDL_VIV_Rep3']}
```

```python
# Remove NaNs from the ENSEMBL column 
no_nan = data.dropna(subset=['ENSEMBL'])
```

```python
filtered = no_nan.groupby('ENSEMBL').mean(numeric_only=True).reset_index()
```

```python
#Calculating mean values of gene expression in each condition for each gene.
for key in SLS:
    filtered[f'SLS_{key}_mean']= filtered[SLS[key]].mean(axis=1)
    print(filtered[f'SLS_{key}_mean'])
```
